```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

# Load data live jika ada, fallback dummy
df = pd.read_csv("../data/jobs_live.csv" if os.path.exists("../data/jobs_live.csv") else "../data/dummy_jobs.csv")
```

# ðŸ“Š Bar Chart â€“ Jumlah Lowongan per Lokasi


```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='location')
plt.title("Jumlah Lowongan per Lokasi")
plt.tight_layout()
plt.savefig("../app/static/img/plot_lokasi.png", dpi=150)
plt.show()
```

# ðŸ“Š Boxplot â€“ Sebaran Gaji per Lokasi


```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='location', y='salary', palette='Set2')
plt.title("Sebaran Gaji per Lokasi")
plt.tight_layout()
plt.savefig("../app/static/img/plot_gaji.png", dpi=150)
plt.show()
```
